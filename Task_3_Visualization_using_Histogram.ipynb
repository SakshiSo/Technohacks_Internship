```python
import pandas as pd
import matplotlib.pyplot as plt
import io
```

```python
data = pd.read_csv(io.BytesIO(uploaded['Iris (1).csv']))
data
```

```python
data.dtypes
```

```python
plt.hist(data['SepalLengthCm'], bins = 10, edgecolor = 'black', alpha = 0.7)
plt.xlabel('SepalLengthCm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length in Cm')
plt.show()
```

```python
plt.hist(data['SepalWidthCm'], bins = 10, edgecolor = 'black', alpha = 0.7)
plt.xlabel('SepalWidthCm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width in Cm')
plt.show()
```

```python
plt.hist(data['PetalLengthCm'], bins = 10, edgecolor = 'black', alpha = 0.7)
plt.xlabel('PetalLengthCm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length in Cm')
plt.show()
```

```python
plt.hist(data['PetalWidthCm'], bins = 10, edgecolor = 'black', alpha = 0.7)
plt.xlabel('PetalWidthCm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width in Cm')
plt.show()
```

```python
species_counts = data['Species'].value_counts()
plt.bar(species_counts.index, species_counts.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Count of Each Species')
plt.show()
```
